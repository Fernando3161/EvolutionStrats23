```python
# Analysis of the function convergence
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

method_names = ["sphere","rosenbrock", "rastring"]
METHOD = 1 # ["sphere","rosenbrock", "rastring"]

```

```python
# Evaluation of convergence vs. Iteration runs for 3 Benchmark functions
import os
import pandas as pd
LAMDA = 40
res = {}
for METHOD in range(3):
    csv_name = os.path.join(os.getcwd(),"results", f"one_lamda_sa_M_{METHOD}_L_{LAMDA}.csv")
    df = pd.read_csv(csv_name, header=0, index_col=0)
    fit = df["best_fit"]
    res[method_names[METHOD]]=fit

df_res = pd.DataFrame.from_dict(res)

sns.set_style("darkgrid")
sns.set(font = "arial")

fig,ax= plt.subplots(figsize = (10,6))
#ax= axs[0]
df_res.plot(ax=ax)
#df_sigmas["sigmas"].plot(ax=ax)
ax.set_xlabel("Iteration")
ax.set_ylabel("Best Fitness")
ax.set_yscale("log")
#ax.set_ylim(bottom = min(res["rosenbrock"]))
ax.set_title("Convergence for Benchmark Algorithms", fontsize= 14, fontweight = "bold")
fig.savefig("results/convergence.png", dpi=600)
```
