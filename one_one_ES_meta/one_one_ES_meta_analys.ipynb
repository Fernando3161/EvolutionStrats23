Analysis of the Rechenberg optimization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

method_names = ["sphere","rosenbrock", "rastring"]
```

```python
METHOD = 1 # ["sphere","rosenbrock", "rastring"]

csv_name = f"rechenberg_{METHOD}.csv"
df_sigmas = pd.read_csv(csv_name, header=0, index_col=0)
df_sigmas
```

```python
for METHOD in [0,1,2]:# ["sphere","rosenbrock", "rastring"]
    csv_name = f"rechenberg_{METHOD}.csv"
    df_sigmas = pd.read_csv(csv_name, header=0, index_col=0)
    df_sigmas

    sns.set_style("darkgrid")
    sns.set(font = "arial")

    fig,axs = plt.subplots(1,3, figsize = (15,6))
    ax= axs[0]
    ax.scatter(x= df_sigmas["sigma_iter"], y = df_sigmas["sigmas"])
    #df_sigmas["sigmas"].plot(ax=ax)
    ax.set_xlabel("Iteration")
    ax.set_ylabel("sigma")
    ax.set_title("Sigma value")

    #fig,ax = plt.subplots()
    ax= axs[1]
    ax.scatter(x= df_sigmas["sigma_iter"], y = df_sigmas["medians"])
    ax.set_xlabel("Iteration")
    ax.set_ylabel("Median")
    ax.set_title(" Median value")

    ax= axs[2]
    ax.scatter(x= df_sigmas["sigmas"], y = df_sigmas["medians"], c=df_sigmas["sigma_iter"] )
    ax.set_xlabel("Sigmas")
    ax.set_ylabel("Median")
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_title("Median vs Sigma - LogLog")
    fig.suptitle("Meta Results - " + method_names[METHOD].capitalize(), fontsize= 14, fontweight = "bold")
    fig.savefig(f"Meta_{METHOD}.png")
```

```python
#fig,ax = plt.subplots()
ax= axs[2]
ax.scatter(x= df_sigmas["sigmas"], y = df_sigmas["medians"])
ax.set_xlabel("Iteration (improved)")
ax.set_ylabel("Median")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_title("Rechenberg Results - Median vs Sigma")
```
