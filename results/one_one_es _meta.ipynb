Results evaluation for the (1 + 1) - ES

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")
sns.set_style("darkgrid")
sns.set(font = "arial")

_mu = "\u03BC"
_lam = "\u03BB"
_sig = "\u03C3"
```

Compare convergence of the functions

```python

df_r = pd.DataFrame()
FUNC_NAMES = ["Sphere", "Rosenbrock", "Rastrigin"]
fig, axs = plt.subplots(1, 3, figsize=(15, 6))
fig.suptitle(f"( 1 + 1 - ES" + "\n" + "META Evolution - Medians",
             fontweight="bold")
i = 0
for n in [2, 5, 10]:
    ax = axs[i]
    i += 1
    df_r = pd.DataFrame()
    for f in [0, 1, 2]:
        path = os.path.join(os.getcwd(), "one_one_es_meta")
        filename = f"one_one_es_meta_F_{f}_n_{n}.csv"
        path = os.path.join(path,filename)
        df = pd.read_csv(path)
        df_r[FUNC_NAMES[f]] = df["medians"]

    ax = sns.lineplot(data=df_r, ax=ax)
    ax.set_yscale("log")
    ax.set_xlabel("Meta - Iteration")
    ax.set_ylabel("Median Value")
    ax.set_title(f"Median Fitness Convergence, N = {n}", fontweight="bold")
```

Analyse Rechenberg Evolution

```python

df_r = pd.DataFrame()
FUNC_NAMES=["Sphere","Rosenbrock", "Rastrigin"]
fig,axs = plt.subplots(1,3, figsize = (15,6))
fig.suptitle(f"( 1 + 1) - ES - Rechenberg Optimization",
                  fontweight= "bold")
i=0
for n in [2,5,10]:
    ax= axs[i]
    i+=1
    df_r = pd.DataFrame()
    for f in [0,1,2]:
        path = os.path.join(os.getcwd(), "one_one_es", f"one_one_es_F_{f}_n_{n}_rech.csv")
        df = pd.read_csv(path)
        df_r[FUNC_NAMES[f]] = df["fitness"]
        
    ax = sns.lineplot(data=df_r, ax= ax)
    ax.set_yscale("log")
    ax.set_ylim(bottom=1e-10)
    ax.set_xlabel("Iteration")
    ax.set_ylabel("Fitness Value")
    ax.set_title(f"Fitness Convergence, N = {n}", fontweight = "bold")
```
